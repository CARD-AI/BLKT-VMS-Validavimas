```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Read the CSV file into a DataFrame
# Replace 'confusion_matrix.csv' with your file path
df = pd.read_csv('../output_roberta_rc3/model/confusion_matrix.csv', index_col=0)  # index_col=0 if first column has labels

# 2. Convert to NumPy array if needed
cm = df.values  # confusion matrix as a NumPy array
labels = df.columns  # class labels from header

# 3. Plot the confusion matrix
plt.figure(figsize=(16, 12))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix (RoBERTa RC3)')

# 4. Add tick marks and labels
tick_marks = np.arange(len(labels))
plt.xticks(tick_marks, labels, rotation=45)
plt.yticks(tick_marks, labels)

# 5. Annotate cells with values
thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, format(cm[i, j], 'd'),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.tight_layout()
plt.savefig('confusion_matrix_roberta_RC3.png')  # Save the plot as a PNG file
plt.show()
```
